```python
import pandas as pd
import requests
import bs4
```

```python
wikiurl="https://en.wikipedia.org/wiki/List_of_cities_in_India_by_population"
response=requests.get(wikiurl)
print(response.status_code)
```

```python
root = bs4.BeautifulSoup(response.text, 'html.parser')
tom_movie=root.find('table',{'class':"wikitable"})
tom_movie
```

```python
columns=[th.text for th in tom_movie.find('tr').find_all('th')]

finalColumns = [columns.replace('\n', '') for columns in columns]

finalColumns
```

```python
trs = tom_movie.find_all('tr')
rows = list()
for tr in trs:
    rows.append([td.text for td in tr.find_all('td')])

print (rows)
```

```python
Final=pd.DataFrame(data=rows, columns=finalColumns)

Final_clean=Final.replace({'\n':''},regex=True)

Final_clean
```
